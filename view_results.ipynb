```python
import pandas as pd
import os 
from pathlib import Path
```

```python
files = []
folder = 'results/rf'
for file in os.listdir(folder):
    files.append(file)

sorted(files,key=len)
```

```python
def pretty_print(file,folder):
    df = pd.read_csv(folder + '/' + file, index_col=[0])
    with pd.option_context('display.max_colwidth', None,'display.max_rows', None):
        display(df.sort_values(by='rank_test_score')[['params','mean_test_score','std_test_score','mean_train_score','std_train_score']])
```

```python
pretty_print('maxdepth,maxfeatures,maxsamples,maxleafnodes,criterion,estimators,samplingstrategy',folder)
```
